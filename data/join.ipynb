```python
import pandas as pd

# Chargement (adapte les noms de fichiers si besoin)
df_menage = pd.read_csv("code_menage.csv")
df_volume = pd.read_csv("volume_de_marché.csv")

# Création de la clé sur 2 chiffres pour 'code_menage'
# On s'assure que c'est du texte, puis on prend les 2 premiers caractères
df_menage['dept_key'] = df_menage['Code_Commune'].astype(str).str[:2]

# Création de la clé sur 2 chiffres pour 'volume_de_marché'
# On utilise la colonne 'code_officiel_commune' (ou code_postal selon ton besoin)
df_volume['dept_key'] = df_volume['code_officiel_commune'].astype(str).str[:2]
```

```python
# On fait une jointure (merge)
# 'left' permet de garder toutes les lignes de ta table principale (volume de marché)
df_final = pd.merge(df_volume, df_menage, on='dept_key', how='left')

# Optionnel : supprimer la colonne temporaire si elle ne te sert plus
df_final = df_final.drop(columns=['dept_key'])

# Afficher le résultat
df_final.head()
```

```python
df_final.to_csv("data_clean_final.csv", index=False)
```
