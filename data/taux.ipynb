```python
import pandas as pd
```

```python
# 1. Charger ton fichier (csv, excel, etc.)
df = pd.read_excel("taux_voiture_menages.xlsx")
df.head()
```

```python


# 2. Définir la liste des départements d'Île-de-France (en chaînes de caractères)
idf_deps = ['75', '77', '78', '91', '92', '93', '94', '95']
```

```python
# On garde toutes les lignes à partir de l'indice 3 jusqu'à la fin
df = df.iloc[3:]

# On réinitialise les index pour que ça reparte de 0
df = df.reset_index(drop=True)

df.head()
```

```python
# On remplace 'Ancien Nom' par 'Departement', etc.
df = df.rename(columns={
    'Équipement automobile des ménages en 2022 : comparaisons départementales': 'code_Commune', 
    'Unnamed: 1': 'departement',
    'Unnamed: 2': '',
    'Unnamed: 3': 'taux_equipement_menages',
    'Unnamed: 4': 'part_menages_voiture',
    'Unnamed: 5': 'part_menages_voitures',

})

df.head()
```

```python
# 3. Filtrer les données
# On suppose que ta colonne s'appelle 'code_postal' ou 'departement'
# On prend les 2 premiers chiffres du code postal pour comparer
df_idf = df[df['Code_Commune'].astype(str).str.startswith(tuple(idf_deps))]
```

```python

# 4. Vérifier le résultat
print(f"Nombre de lignes restantes : {len(df_idf)}")
df_idf.head()
```

```python
# 1. On définit la liste des départements d'Île-de-France
# Note : on les met entre guillemets car ce sont souvent des textes (surtout pour le 75, 77, etc.)
idf = ['75', '77', '78', '91', '92', '93', '94', '95']

# 2. On ne garde que les lignes où la colonne 'Departement' est dans notre liste
df = df[df['Code_Commune'].astype(str).isin(idf)]

# 3. On réinitialise l'index pour que ça reparte de 0, 1, 2...
df = df.reset_index(drop=True)

# 4. On vérifie le résultat
print(f"Il reste {len(df)} lignes pour l'Île-de-France.")
df.head()
```

```python
df.to_csv("code_menages.csv", index=False)
```
